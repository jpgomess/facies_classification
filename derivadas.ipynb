```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import segyio
import os
```

```python
file = '/home/gaia/jpedro/Seismic_data_w_null.sgy'

f3_seismic = segyio.open(file)
```

```python
img = f3_seismic.iline[f3_seismic.ilines[0]]
img = img.T

fig, ax = plt.subplots(3, 1, figsize=(15,21))

vmin = -4000
vmax = 4000

ax[0].imshow(img, cmap='gray_r', vmin=vmin, vmax=vmax)
ax[0].set_title('Default')
ax[1].imshow(np.diff(img) / 25, cmap='gray_r', vmin=vmin, vmax=vmax)
ax[1].set_title('Horizontal')
ax[2].imshow((np.diff(img.T) / 0.004).T, cmap='gray_r', vmin=vmin, vmax=vmax)
ax[2].set_title('Vertical')
```
